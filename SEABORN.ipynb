```python
# SEABORN DATA VISUALISATION LIBRARY

# import warnings
# warnings.filterwarnings("ignore") # used to ignore warnings

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(sns.__version__)
```

```python
# built-in datasets of seaborn
sns.get_dataset_names()
```

```python
dataset = sns.load_dataset("iris")
dataset
```

```python
# Line Graph
# lineplot()
sns.lineplot(x="sepal_length",y="sepal_width",data=dataset)
plt.title("Lineplot In Seaborn")
plt.show()
```

```python
# adding style to seaborn
plt.figure(figsize=(10,5))
sns.set_style("whitegrid", {"grid.color":"lightblue","grid.linestyle":"--"}) # -- dashed : dotted - solid
sns.lineplot(x="sepal_length",y="sepal_width",data=dataset)
plt.title("Lineplot In Seaborn", color="darkblue", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
# scatterplot()
sns.scatterplot(x="Attack", y="Defense", data=pokemon, color="hotpink")
plt.title("Scatter Plot in Seaborn")
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
# scatterplot()
sns.scatterplot(x="Attack", y="Defense",hue="Generation", data=pokemon)
plt.title("Scatter Plot in Seaborn")
plt.show()
```

```python
color = ['r',"y","b","g","c","m"]
sns.scatterplot(x="Attack", y="Defense",hue="Generation",palette=color, data=pokemon)
plt.title("Scatter Plot in Seaborn")
plt.show()
```

```python
# lmplot()
# used to generate a scatter plot with a regression line

color = ['r',"y","b","g","c","m"]
sns.lmplot(x="Attack", y="Defense",hue="Generation",palette=color, data=pokemon)
plt.title("Scatter Plot in Seaborn")
plt.show()
```

```python
# countplot()
# used to generate a bar graph

plt.figure(figsize=(15,5))
sns.countplot(x="Type 1", data=pokemon)
plt.title("Countplot in Seaborn")
plt.show()
```

```python
pokemon["Type 1"].value_counts()
```

```python
# Heatmap
# A heatmap is a colored table where colors represent data values,
# helping to quickly identify high, low, positive, or negative values.

'''
Common places where heatmaps are used:

1. Correlation Analysis (Very common in data science)

To show relationships between variables

Example:

Study hours vs marks

Features in machine learning datasets

Helps spot strong positive or negative relationships quickly

2. Academic Performance / Education

Student scores across subjects

Attendance vs performance

Identifying strong and weak areas

3. Website & App Analytics

To see where users click or spend time

Mouse movement or scrolling behavior

Dark areas = more user activity

4. Business & Sales Analysis

Sales performance by region and time

Product demand across months

High and low revenue zones

'''

data = {
    'Hours_Studied':[2,3,4,5,6,7,8],
    'Marks_Scored':[40,50,60,70,80,90,100],
    'TV_Watched':[10,9,8,6,7,5,3]
}

df = pd.DataFrame(data)
df
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, cmap="Blues", annot=True) # cmap-->"viridis","coolwarm","Blues",etc.
plt.title("Heatmap")
plt.show()
```
