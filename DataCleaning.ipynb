```python
# Working with Missing Data in Pandas
# Missing Data can occur when no information is provided for one or more
# items or for a whole unit. Missing Data is a very big problem in a real-life scenarios.
# Missing Data can also refer to as NA(Not Available) values in pandas.
# In DataFrame sometimes many datasets simply arrive with missing data,
# either because it exists and was not collected or it never existed.
# For Example, Suppose different users being surveyed may choose not to share their income,
# some users may choose not to share the address in this way many datasets went missing.

# Dataset 1 : https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv

# Dataset 2 : https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

ufo = pd.read_csv("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/ufo.csv")
ufo.head()
```

```python
ufo.info()
```

```python
ufo.isnull().sum()
```

```python
# remove missing values
# dropna()
# not advisable to be used with large number of missing values

ufo.dropna().shape
```

```python
# fillna()
# Fill NA/NaN Values using the specified method

ufo["City"].fillna(value="CityNotMentioned", inplace=True)

ufo["Colors Reported"].fillna(value="Unidentified Colors", inplace=True)

ufo["Shape Reported"].fillna(value="Unidentified Shape", inplace=True)
```

```python
data1 = pd.read_csv("https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/sample11.csv")
data1
```

```python
# Fill Missing Values By Taking Up A Mean
mean_val_marks = data1["Marks"].mean()
data1["Marks"].fillna(value=mean_val_marks, inplace=True)
data1
```

```python
mean_val_percentage = data1["Percentage"].mean()
mean_val_percentage = round(mean_val_percentage, 1)
data1["Percentage"].fillna(value=mean_val_percentage, inplace=True)
data1
```

```python
# Remove Duplicate Records
# Handle Duplicate Data
# Dealing With Duplicate Data

# Create a DataFrame
data = {
      'StudentName': ['Mark', 'Ali', 'Bob', 'John', 'Johny', 'Mark'],
      'Score': [45, 65, 76, 44, 39, 45]
}
df = pd.DataFrame(data)
df
```

```python
# check for duplicate records
print(df.duplicated().sum())
```

```python
df[df.duplicated()]
```

```python
df[df["StudentName"]=="Mark"]
```

```python
# dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df
```
